```python
import time
import json
import numpy as np

from AStar import *


def get_current_t_ms():
    return int(time.time() * 1000)
```

```python
json_file_path = "your_obstacle.json"
with open(json_file_path, 'r') as j:
    contents = json.loads(j.read())
data = json.loads(contents["data"])
```

```python
results = dict()

for n in range(10, 100):
    print("n =", n)
    results[n] = []
    for _ in range(10):
        a_star = AStar(
            cols=int(n),  
            rows=int(n), 
            start=[0, 0], 
            end=[n-1, n-1], 
            obstacle_ratio=20, 
            obstacle_list=False
#             obstacle_ratio=False, 
#             obstacle_list=data
        )

        start_t_ms = get_current_t_ms()
        final_path = a_star.main()
        t_ms = get_current_t_ms() - start_t_ms
        results[n].append(t_ms)


# if len(final_path) > 0:
#     print("The way found!!!")
#     for i in range(len(final_path)):
#         print(final_path[i].x, final_path[i].y)
# else:   
#     print("There is no legal way...You can decrease obstacle ration (default 20)")
```
